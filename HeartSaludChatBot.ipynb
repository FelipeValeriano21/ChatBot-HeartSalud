### Passo 02: Coleta dos dados

```python
nlp = spacy.load('pt_core_news_sm')
```

```python
import urllib.request
from bs4 import BeautifulSoup as bs

def indexarDados(urls):
    conteudo_completo = ""

    for url in urls:
        try:
            link = urllib.request.urlopen(url)
            conteudo = link.read()
            conteudo_html = bs(conteudo, 'lxml')


            paragrafos = conteudo_html.find_all('p')
            conteudo = ''

            for p in paragrafos:
                conteudo += p.text + "\n"

            conteudo_completo += conteudo + "\n\n"
        except Exception as e:
            print(f"Erro ao acessar a URL {url}: {e}")

    return conteudo_completo


urls = [
        'https://pt.wikipedia.org/wiki/Cora%C3%A7%C3%A3o',
        'https://pt.wikipedia.org/wiki/Sistema_circulat%C3%B3rio',
        'https://pt.wikipedia.org/wiki/Cora%C3%A7%C3%A3o_artificial',
        'https://pt.wikipedia.org/wiki/Transplante_de_cora%C3%A7%C3%A3o'

]

conteudo_combined = indexarDados(urls)

print(conteudo_combined)
```

### Passo 04: Etapa de Pré-processamento

```python
def preprocessamento(texto):
  texto = re.sub(r"https?://[A-Za-z0-9./]+", ' ', texto)
  documento = pln(texto)
  lista = []
  for token in documento:
    lista.append(token.lemma_)
  lista = [palavra for palavra in lista if palavra not in stop_words and palavra not in string.punctuation]
  lista = ' '.join([str(elemento) for elemento in lista if not elemento.isdigit()])
  return lista
```

```python
lista_sentencas_preprocessada = []
for i in range(len(lista_sentencas)):
  lista_sentencas_preprocessada.append(preprocessamento(lista_sentencas[i]))
```

```python
textos_boas_vindas_entrada = ('hey', 'olá', 'opa', 'oi', 'eae', 'salve')
textos_boas_vindas_respostas = ('olá me faça uma pergunta de coração ❤️', 'Bora conversar sobre ❤️?', 'O que você manda ❤️?', 'pergunta ai sobre coração ❤️')
```

```python
textos_despedidas_entrada = ('tchau', 'adeus', 'até mais', 'até logo', 'falou', 'até breve')
textos_despedidas_respostas = ('Até mais! Qualquer coisa estou aqui', 'Até mais colega!', 'Espero que tenha aproveitado a experiência!', 'Até breve, valeu!')
```

```python
def responder_saudacao(texto):
  for palavra in texto.split():
    if palavra.lower() in textos_boas_vindas_entrada:
      return random.choice(textos_boas_vindas_respostas)
```

```python
def responder_despedida(texto):
  for palavra in texto.split():
    if palavra.lower() in textos_despedidas_entrada:
      return random.choice(textos_despedidas_respostas)
```

```python
inicio_de_respostas = (
    'Boa pergunta!',
    'Deixe me ver, ',
    'Interessante questão, ',
    'Bom, ',
    'Então, ',
    'Ótima dúvida!',
    'Vamos lá, ',
    'Excelente pergunta!',
    'Certo, ',
    'Vamos analisar isso, '
)
```

```python
def responder(texto_usuario):
  resposta_chatbot = ''
  lista_sentencas_preprocessada.append(texto_usuario)

  tfidf = TfidfVectorizer()
  palavras_vetorizadas = tfidf.fit_transform(lista_sentencas_preprocessada)

  similaridade = cosine_similarity(palavras_vetorizadas[-1], palavras_vetorizadas)

  indice_sentenca = similaridade.argsort()[0][-2]
  vetor_similar = similaridade.flatten()
  vetor_similar.sort()
  vetor_encontrado = vetor_similar[-2]

  if (vetor_encontrado == 0):
    resposta_chatbot = '\nDesculpe, ainda não sei essa pergunta! Que tal uma nova pergunta de coração?'
    return resposta_chatbot
  else:
    resposta_chatbot = resposta_chatbot + lista_sentencas[indice_sentenca]
    return resposta_chatbot
```

### Passo 07: Ultilizar o HeartSalud

```python
import random  # Importando o módulo random

continuar = True
print('\n🫀Chatbot:\n')
print('Olá, sou o HeartSalud! gosto de falar sobre coração, me pergunte algo:\n ')
print('Use a documentação em caso de dificuldade\n ')
while continuar:
    texto_usuario = input()
    texto_usuario = texto_usuario.lower()

    if texto_usuario not in textos_despedidas_entrada:
        resposta_saudacao = responder_saudacao(texto_usuario)

        if resposta_saudacao is not None:
            print('---------------------------------------------------------------------------')
            print('\n🫀Chatbot:\n')
            print( resposta_saudacao + '\n')
            print('---------------------------------------------------------------------------')
        else:
            print('---------------------------------------------------------------------------')
            print('\n🫀Chatbot:\n')
            print(random.choice(inicio_de_respostas) + " " + responder(preprocessamento(texto_usuario)) + " " + '\n')
            print('---------------------------------------------------------------------------')
            lista_sentencas_preprocessada.remove(preprocessamento(texto_usuario))
    else:
      continuar = False
      despedida = random.choice(textos_despedidas_respostas)  # Corrigido de ramdom para random
      print('\n🫀Chatbot: ' + despedida)

```
